***Zadanie 1.***    
Wyznacz numerycznie wartość całki:   
$$
\int\limits_{a}^{b}\sin^2(\cfrac{1}{x-1.01})dx 
$$

Wykonaj następujące czynności:

1. Czy możliwe jest całkowanie podanej funkcji na dowolnym przedziale? Swoją odpowiedź uzasadnij.
2. Oblicz całkę numerycznie na przedziale $a =0$, $b=1$ przy pomocy następujących metod:  
    1. Metoda Prostokątów
    2. Metoda Trapezów - [trapezoid](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.trapezoid.html#scipy.integrate.trapezoid)
    3. Metoda Kwadratur - [quad](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html#scipy.integrate.quad)

```python
#funckja 1 

f = lambda x: np.sin(1/(x-1.01))**2
a = 0 
b = 1 

resquad, errorquad = sp.integrate.quad(f, a, b)
print(resquad)


resquadrature1, errorquadrature1 = sp.integrate.quadrature(f, a, b,maxiter=1)
print(resquadrature1)

resquadrature10, errorquadrature10 = sp.integrate.quadrature(f, a, b,maxiter=10)


resquadrature50, errorquadrature50 = sp.integrate.quadrature(f, a, b,maxiter=50)


resquadrature100, errorquadrature100 = sp.integrate.quadrature(f, a, b,maxiter=100)


resquadrature1000, errorquadrature1000 = sp.integrate.quadrature(f, a, b,maxiter=1000)

resquadrature3000, errorquadrature3000 = sp.integrate.quadrature(f, a, b,maxiter=3000)

x = [1, 10, 50, 100, 1000, 3000]
res_quad = [resquad,resquad,resquad,resquad,resquad,resquad]
res_quadrature = [resquadrature1,resquadrature10,resquadrature50,resquadrature100,resquadrature1000,resquadrature3000]

plt.plot(x, res_quad, 'g--',  label = 'Wynik quad')
plt.xlabel('Argumenty')
plt.ylabel('Resultat')
plt.legend()
plt.title('Wyniki dla metody quad')
plt.show()

plt.scatter(x, res_quadrature, label = 'Wynik quadrature')
plt.xlabel('Dla różnych n')
plt.ylabel('Resultat')
plt.legend()
plt.title('Wyniki dla metody quadrature')
plt.show()

```

WNIOSKI Z 1 ZADANIA 

Metoda quad nie posiada możliwości zmian liczby iteracji, więc wszystkie wyniki dla niej są jednakowe. Inne metody posiadają różnice przy wyborze różnych n. Metoda trapezoidu posiada dokładność dx jednak w tym jednym wykresie, porównującym wszystkie metody uwzględniłem jedynie od najmniejszej do największej dokładości, aby zwizualizować. Nie trudno się było domyślić, że czas będzie wzrastał z dokładnoścą. Jednak przy trapezoidzie nie jest to aż tak widoczne. Również nietrudno było się domyślić, że jeżeli będzie więcej iteracji to bedzie lepsza dokładność. Metoda prostokątów od początku ma dosyć dokładne wyniki, nie można tak powiedzieć o trapezoidzie oraz quadrature.

Metoda quad nie posiada możliwości zmian liczby iteracji, więc wszystkie wyniki dla niej są jednakowe. Inne metody posiadają różnice przy wyborze różnych n. Metoda trapezoidu posiada dokładność dx jednak w tym jednym wykresie, porównującym wszystkie metody uwzględniłem jedynie od najmniejszej do największej dokładości, aby zwizualizować. Nie trudno się było domyślić, że czas będzie wzrastał z dokładnoścą. Jednak przy trapezoidzie nie jest to aż tak widoczne. Również nietrudno było się domyślić, że jeżeli będzie więcej iteracji to bedzie lepsza dokładność. Metoda prostokątów od początku ma dosyć dokładne wyniki, nie można tak powiedzieć o trapezoidzie oraz quadrature.

WNIOSKI OGÓLNE

W ogólnym wnioski to najłatwiej jest używać metody quad. Jest ona najdokładniejsza, ponieważ sama dobiera sobie liczbe iteracji. Jest ona również zaskakująco szybka. Następną metodą byłaby quadrature, ponieważ jest ona również bardzo dokładna, dla niektórych funckji również dla małych iteracji, jednakże czas trwania tej funckji jest najdłuższy. Do metody prostokątów potrzebna jest pętla co myśle, że również przełoż się na długi czas, a nie jest aż tak dokładna. Metoda trapezoidu jeżeli się wybierze odpowiednio mały krok to też może być bardzo dobra bo mało czasu potrzebuje.
