***Zadanie 1***   

Wykorzystując funkcję *random_matrix_by_egval* wygenerować losowe macierze dla zadanych zestawów wartości własnych:

$\lambda_{1} = \{ 0 , 1 , 2 , ... , 20 \}$

$\lambda_{2} = \{ 10^{-10} , 10^{-9} , ... , 10^{-1} , 1 \}$

$\lambda_{3} = \{ j , -j , 1+j , 1-j \}$

Następnie dla wygenerowanych macierzy oblicz wartości własne za pomocą funkcji *eigvals* i wyrysuj je na wykresie.

```python

lambda1 = np.linspace(0,20,21)
lambda2 = np.array([10**-10, 10**-9, 10**-8, 10**-7, 10**-6, 10**-5, 10**-4, 10**-3, 10**-2, 10**-1, 10**0 ])
lambda3 = np.array([0+1.j , 0-1.j , 1+1.j , 1-1.j], dtype = complex)
#print(lambda1)
#print(lambda2)
#print(lambda3)
#print(lambda3[3])

macierz1 = main.random_matrix_by_egval(lambda1)
#print(macierz1)
macierz2 = main.random_matrix_by_egval(lambda2)
#print(macierz2)
macierz3 = main.random_matrix_by_egval(lambda3)
#print(macierz3)
wartosci1 = np.linalg.eigvals(macierz1)
print(wartosci1)
wartosci1_sort = np.sort(wartosci1)
wartosci2 = np.linalg.eigvals(macierz2)
print(wartosci2)
wartosci2_sort = np.sort(wartosci2)

wartosci3 = np.linalg.eigvals(macierz3)
print(wartosci3)
wartosci3_sort = np.sort(wartosci3)

plt.scatter(lambda1, wartosci1_sort)
plt.title("Wykres wartości własnych dla lambda1")
plt.xlabel("lambda1")
plt.ylabel("wartosci1sort")
plt.grid()
plt.show()

plt.scatter(lambda2, wartosci2_sort)
plt.title("Wykres wartości własnych dla lambda2")
plt.xscale('log')
plt.xlabel("lambda2")
plt.ylabel("wartosci2sort")
plt.grid()
plt.show()

plt.scatter(np.real(wartosci3_sort) ,np.imag(wartosci3_sort),   label = 'wyliczone')
plt.scatter(np.real(lambda3), np.imag(lambda3), label = ' wyjściowe' )
plt.title("Wykres wartości własnych dla lambda3")
plt.xlabel("real")
plt.ylabel("imag")
plt.legend()
plt.grid()
plt.show()

blad = main.L_inf(lambda1,wartosci1_sort)
print(blad)

blad = main.L_inf(lambda2,wartosci2_sort)
print(blad)



```

Wykorzystując zestawy wartości własnych z zadania pierwszego wygeneruj macierze funkcją *random_matrix_by_egval* (lub uzyj wygenerowanych wcześniej macierzy). 

Następnie do otrzymanych wyników dodaj macierz $B$, gdzie $B = \alpha I$, przy czym $\alpha$ równa się ilości liter w Twoim nazwisku.

Wypisz wartości własne otrzymanej w ten sposób macierzy. Porównaj je z pierwotnie wykorzystanymi.

```python
#maceirz1 , mceirz2 , macierz3 

alpha = len('Dulak')
#print(alpha)
B1 = np.multiply(alpha,np.eye(macierz1.shape[0]))
B2 = np.multiply(alpha,np.eye(macierz2.shape[0]))
B3 = np.multiply(alpha,np.eye(macierz3.shape[0]))
#print(B1)
#print(B2)
#print(B3)
macierz21 = macierz1 + B1 
#print(macierz21)
macierz22 = macierz2 + B2 
#print(macierz22)
macierz23 = macierz3+ B3 
#print(macierz23)

wartosci21 = np.linalg.eigvals(macierz21)
print(wartosci21)
wartosci21_sort = np.sort(wartosci21)
wartosci22 = np.linalg.eigvals(macierz22)
print(wartosci22)
wartosci22_sort = np.sort(wartosci22)
wartosci23 = np.linalg.eigvals(macierz23)
print(wartosci23)
wartosci23_sort = np.sort(wartosci23)



plt.scatter(lambda1, wartosci21_sort, label = 'wyliczone')
plt.scatter(lambda1 , lambda1 , label = ' wyjściowe')
plt.title("Wykres wartości własnych dla lambda1")
plt.xlabel("lambda1")
plt.ylabel("wartosci21sort")
plt.legend()
plt.grid()
plt.show()

plt.scatter(lambda2, wartosci22_sort,  label = 'wyliczone')
plt.scatter(lambda2, lambda2 , label = ' wyjściowe' )
plt.title("Wykres wartości własnych dla lambda2")
plt.xscale('log')
plt.xlabel("lambda2")
plt.ylabel("wartosci22sort")
plt.legend()
plt.grid()
plt.show()


plt.scatter( np.real(wartosci23_sort) ,np.imag(wartosci23_sort),   label = 'wyliczone')
plt.scatter(np.real(lambda3), np.imag(lambda3), label = ' wyjściowe' )
plt.title("Wykres wartości własnych dla lambda3")
plt.xlabel("real")
plt.ylabel("imag")
plt.legend()
plt.grid()
plt.show()

blad = main.L_inf(lambda1,wartosci21_sort)
print(blad)


```
