***Zadanie 6.***  
Dla głównej funkcji oraz przedziału zdefiniowanego z zadania 1 znajdź miejsca zerowe przy użyciu:

* metody bisekcji
* metody siecznych

z tolerancją $10^{-10}$

Zbadaj dokładność(względem rozwiązania z zadania 5) i czas obliczeń metod w zależności od liczby iteracji. Wyniki przedstaw na wykresach. 

```python
x = np.linspace(0,100,1000)
zero = np.zeros(x.shape)
#metoda bisekcji 

b1 = main.bisection(10**-10,1,main.fun, 10**-10 ,1)
errb1 = abs(b1[0]-roots.x[1])
b5 = main.bisection(10**-10,1,main.fun, 10**-10 ,5)
errb5 = abs(b5[0]-roots.x[1])
#print(err1)
b10 = main.bisection(10**-10,1,main.fun, 10**-10 ,10)
errb10 = abs(b10[0]-roots.x[1])
b20 = main.bisection(10**-10,1,main.fun, 10**-10 ,20)
errb20 = abs(b20[0]-roots.x[1])
b50 = main.bisection(10**-10,1,main.fun, 10**-10 ,50)
errb50 = abs(b50[0]-roots.x[1])
b100 = main.bisection(10**-10,1,main.fun, 10**-10 ,100)
errb100 = abs(b20[0]-roots.x[1])
biswithiter = [b1,b5, b10, b20, b50, b100]
print(biswithiter)
#Widac ze nieważne ile bym dał więcej iteracji to juz nie ma sensu ponieważ przy 31 iteracji juz jest wartosc wymagana 
onlybis = [b1[0], b5[0], b10[0], b20[0], b50[0], b100[0]]
print(onlybis)
errb = [errb1,errb5, errb10, errb20,errb50, errb100]
print(errb)

saved_timing = %timeit -r 5 -n 10 -o main.bisection(10**-10,1,main.fun, 10**-10 ,1)
timeb1 = saved_timing.average   

saved_timing = %timeit -r 5 -n 10 -o main.bisection(10**-10,1,main.fun, 10**-10 ,5)
timeb5 = saved_timing.average   
    
saved_timing = %timeit -r 5 -n 10 -o main.bisection(10**-10,1,main.fun, 10**-10 ,10)
timeb10 = saved_timing.average

saved_timing = %timeit -r 5 -n 10 -o main.bisection(10**-10,1,main.fun, 10**-10 ,20)
timeb20 = saved_timing.average   

saved_timing = %timeit -r 5 -n 10 -o main.bisection(10**-10,1,main.fun, 10**-10 ,50)
timeb50 = saved_timing.average  

saved_timing = %timeit -r 5 -n 10 -o main.bisection(10**-10,1,main.fun, 10**-10 ,100)
timeb100 = saved_timing.average   

timeb = [timeb1,timeb5, timeb10, timeb20, timeb50, timeb100]
n = [1, 5, 10, 20, 50, 100]
zero = zero + roots.x[1]
plt.scatter(n,onlybis,label='Wyniki dla poszczególnych iteracji')
plt.plot(x,zero,'r--',label='Wartość wyliczona solve')
plt.xlabel('Ilośc iteracji')
plt.ylabel('Wartości')
plt.title('Wyniki działania metody bisekcji dla różnych iteracji')
plt.legend()
plt.show()

plt.scatter(n,errb,label='Błąd')
plt.xlabel('Ilośc iteracji')
plt.ylabel('Wartości błędu')
plt.title('Wyniki działania metody bisekcji dla różnych iteracji - błąd')
plt.legend()
plt.yscale('log')
plt.show()


plt.scatter(n,timeb,label='Czas')
plt.xlabel('Ilośc iteracji')
plt.ylabel('Czas')
plt.title('Wyniki działania metody bisekcji dla różnych iteracji - czas')
plt.legend()
plt.yscale('log')
plt.show()


```

```python
#metoda siecznych 

s1 = main.secant(10**-10,1,main.fun, 10**-10 ,1)
errs1 = abs(s1[0]-roots.x[1])
s5 = main.secant(10**-10,1,main.fun, 10**-10 ,5)
errs5 = abs(s5[0]-roots.x[1])
s10 = main.secant(10**-10,1,main.fun, 10**-10 ,10)
errs10 = abs(s10[0]-roots.x[1])
s20 = main.secant(10**-10,1,main.fun, 10**-10 ,20)
errs20 = abs(s20[0]-roots.x[1])
s50 = main.secant(10**-10,1,main.fun, 10**-10 ,50)
errs50 = abs(s50[0]-roots.x[1])
s100 = main.secant(10**-10,1,main.fun, 10**-10 ,100)
errs100 = abs(s100[0]-roots.x[1])

secwithiter = [s1,s5, s10, s20, s50, s100]
print(secwithiter) 
#Widac ze wystarczy 16 iteracji  
onlysec = [s1[0], s5[0], s10[0], s20[0], s50[0], s100[0]]
print(onlysec)
errs = [errs1,errs5, errs10, errs20,errs50, errs100]
print(errs)


saved_timing = %timeit -r 5 -n 10 -o main.secant(10**-10,1,main.fun, 10**-10 ,1)
times1 = saved_timing.average   

saved_timing = %timeit -r 5 -n 10 -o main.secant(10**-10,1,main.fun, 10**-10 ,5)
times5 = saved_timing.average   
    
saved_timing = %timeit -r 5 -n 10 -o main.secant(10**-10,1,main.fun, 10**-10 ,10)
times10 = saved_timing.average

saved_timing = %timeit -r 5 -n 10 -o main.secant(10**-10,1,main.fun, 10**-10 ,20)
times20 = saved_timing.average   

saved_timing = %timeit -r 5 -n 10 -o main.secant(10**-10,1,main.fun, 10**-10 ,50)
times50 = saved_timing.average  

saved_timing = %timeit -r 5 -n 10 -o main.secant(10**-10,1,main.fun, 10**-10 ,100)
times100 = saved_timing.average   

times = [times1,times5, times10, times20, times50, times100]
n = [1, 5, 10, 20, 50, 100]

plt.scatter(n,onlysec,label='Wyniki dla poszczególnych iteracji')
plt.plot(x,zero,'r--',label='Wartość wyliczona solve')
plt.xlabel('Ilośc iteracji')
plt.ylabel('Wartości')
plt.title('Wyniki działania metody siecznych dla różnych iteracji')
plt.legend()
plt.show()

plt.scatter(n,errs,label='Błąd')
plt.xlabel('Ilośc iteracji')
plt.ylabel('Wartości błędu')
plt.title('Wyniki działania metody siecznych dla różnych iteracji - błąd')
plt.legend()
plt.yscale('log')
plt.show()


plt.scatter(n,times,label='Czas')
plt.xlabel('Ilośc iteracji')
plt.ylabel('Czas')
plt.title('Wyniki działania metody siecznych dla różnych iteracji - czas')
plt.legend()
plt.yscale('log')
plt.show()
```
